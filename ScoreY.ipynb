```python
import pymongo
import numpy as np
import time
from datetime import datetime
import itertools
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client.csi
```

```python
trainEvents = np.load('./TrainTestValCreation/trainEvents.npy')
testEvents = np.load('./TrainTestValCreation/testEvents.npy')
valEvents = np.load('./TrainTestValCreation/valEvents.npy')
```

```python
global_userlist = list(np.load('./global_userlist.npy'))
```

```python
from scipy.sparse import csr_matrix
y = csr_matrix((len(global_userlist), len(global_userlist)), dtype=np.int8).toarray()
```

```python
for e,l in trainEvents:
    user_ids = db.finalHydratedtweets.find({'myeventIndex':int(e)},{'user.id_str':1})
    eUser = []
    Users = []
    for user in user_ids:
        Users.append(int(user['user']['id_str']))
    Users = list(set(Users))
    for user in Users:
        idx = global_userlist.index(user)
        eUser.append(idx)
    userUser = []
    userUser = list(itertools.combinations(eUser, 2))
    for i,j in userUser:
        y[i,j] = y[i,j] + 1
    print (e)
```

```python
np.save('./Dataset1/ytrainset', y)
```

```python
from scipy.sparse import csc_matrix
from scipy.sparse.linalg import svds

score_y = csc_matrix(y, dtype=float)
u, s, vt = svds(score_y, k=50)

s[::-1].sort()
S = np.diag(s)

score_y = np.dot(u,S)
np.save('./score_y_trainset', score_y)
```
